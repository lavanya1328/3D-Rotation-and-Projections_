```python
import numpy as np
import json
```

```python
np.random.seed(42)
total=1000
```

```python
x=np.random.uniform(-400,400,total)
y=np.random.uniform(-400,400,total)
z=np.random.uniform(-400,400,total)
print(x[:3])
print(y[:3])
print(z[:3])
```

```python
points=[]
for nums in range (total):
    points.append({"x":float(x[nums]),
                   "y":float(y[nums]),
                   "z":float(z[nums])})
```

```python
print(points[:4])
```

```python
with open("data.json","w")as file:
    data=json.dump(points,file,indent=4)
print(f"{total} data generated successfully")
```
